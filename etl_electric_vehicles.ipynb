```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ev_per_county_ny_raw.csv')
df
```

```python
df[df['REDC'] == 'New York City'] #Vemos la información disponible para la región de interés
```

```python
#Cambiamos los nombres de county a sus equivalentes en borough, como están en los datasets de taxis
df['County'] = df['County'].replace(['New York County', 'Queens County', 'Kings County', 'Richmond County', 'Bronx County'], ['Manhattan', 'Queens', 'Brooklyn', 'Staten Island', 'Bronx'])
df[df['REDC'] == 'New York City'] #Vemos la información disponible para la región de interés
```

```python
df.describe()
```

```python
df['Light-Duty Vehicles'].sum() #Comprobamos que el total coincide con el Dashboard oficial de PowerBI (fuente de datos)
```

```python
df.info() #Vemos que hay algunos valores nulos
```

```python
#Sabemos que en la fuente original (Dashboard PowerBI) los valores nulos son considerados como 0 y realizamos este cambio
df.replace(np.nan, 0, inplace=True)
df.info()
```

```python
print(df['LD EV Market Share'].unique()) #Empezamos a visualizar y corregir las filas numericas que tienen object dtype.
```

```python
df['LD EV Market Share'] = df['LD EV Market Share'].str.replace('%', '')
df['LD EV Market Share'] = df['LD EV Market Share'].str.replace(',', '.')
df['LD EV Market Share'] = df['LD EV Market Share'].astype(float)
#Cambio de nombre para recordar que el valor está expresado en %
df.rename(columns={'LD EV Market Share': 'LD EV Market Share (%)'}, inplace=True)
```

```python
print(df['LDVs per 1k People'].unique())
print(df['BEVs per 1k People'].unique())
print(df['PHEVs per 1k People'].unique())
print(df['EVs per Level 2 Port'].unique())
```

```python
df['LDVs per 1k People'] = df['LDVs per 1k People'].str.replace(',', '.')
df['LDVs per 1k People'] = df['LDVs per 1k People'].astype(float)

df['BEVs per 1k People'] = df['BEVs per 1k People'].str.replace(',', '.')
df['BEVs per 1k People'] = df['BEVs per 1k People'].astype(float)

df['PHEVs per 1k People'] = df['PHEVs per 1k People'].str.replace(',', '.')
df['PHEVs per 1k People'] = df['PHEVs per 1k People'].astype(float)

df['EVs per Level 2 Port'] = df['EVs per Level 2 Port'].str.replace(',', '.')
df['EVs per Level 2 Port'] = df['EVs per Level 2 Port'].astype(float)
```

```python
#Eliminamos filas 63 y 64 correspondientes a ubicaciones desconocidas.
#Los datos se visualizarán en un mapa por lo que estas filas no serán utilizadas.
df.drop([63,64], axis=0, inplace=True) 
df
```

```python
df.to_csv('ev_per_county_ny.csv', index=False)
```
